## Data Ingestion

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
! uv pip install pypdf
```

```python
loader = PyPDFLoader('D:\\Machine learning\\llmops_series\\data\\Physics 9th Ch 1 Final.pdf')
documents = loader.load()
```

```python
documents[10].page_content # Display the first 500 characters of the first document
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    length_function=len
)
```

```python
text_chunks = text_splitter.split_documents(documents)
```
